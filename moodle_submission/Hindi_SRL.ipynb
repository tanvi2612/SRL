```python
import pandas as pd
import numpy as np
import sys
from sklearn.preprocessing import LabelEncoder
import joblib
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
pd.options.mode.chained_assignment = None
```

```python

dataset = pd.read_csv('./expected.csv') #edit the path to the csv format input file as necessary
# It is required that the field names of the input text mirror that of the example csv provided,i.e expected.csv
print("Dataset provided has the following as its head")
print(dataset.head())
```

```python
# MUST HANDLE THE LABELS WITH THE LABELENCODER
phase1=joblib.load('./models/binary_model46.pkl')
phase2=joblib.load('./models_2/binary_model46.pkl')

print("Starting phase 1")
ph1_params=[]
for i in range(0,300):
    ph1_params.append('dim_{}'.format(i))
ph1_params.append('dependency')
ph1_params.append('dependency-head')
work_set=dataset[ph1_params]

le1=LabelEncoder()
le2=LabelEncoder()

work_set['dependency']=le1.fit_transform(work_set['dependency'])
work_set['dependency-head']=le2.fit_transform(work_set['dependency-head'])
work_set
```

```python
ph1_predictions=phase1.predict(work_set)

```

```python
ph1_predictions
```

```python
work_set['Label']=ph1_predictions
print("Finished phase 1")

print("Starting phase 2")
work_set2=work_set[work_set['Label']==1]
work_set2.head()
```

```python
work_set2=work_set2.drop(['Label'],axis=1)
work_set2.head()
```

```python
srl_labels=LabelEncoder()
srl_labels.fit(dataset['srl'])
ph2_predictions=phase2.predict(work_set2)
print("Finished phase 2")
srl_labels.inverse_transform(ph2_predictions)
```
